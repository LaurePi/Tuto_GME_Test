```python
import numpy as np
import matplotlib.pyplot as plt
import ipywidgets as widgets
from IPython.display import display

def plot_demo(freq=1.0, amplitude=1.0, noise=0.0):
    x = np.linspace(0, 10, 400)
    rng = np.random.default_rng(123)
    y = amplitude * np.sin(freq * x) + rng.normal(0, noise, size=len(x))

    fig, ax = plt.subplots(figsize=(8, 4))
    ax.plot(x, y)
    ax.set_title("Widget test - sinusoïde")
    ax.set_xlabel("x")
    ax.set_ylabel("y")
    ax.grid(True)
    plt.show()

w_freq = widgets.FloatSlider(
    value=1.0, min=0.5, max=5.0, step=0.1,
    description="Fréquence", continuous_update=False
)

w_amp = widgets.FloatSlider(
    value=1.0, min=0.1, max=3.0, step=0.1,
    description="Amplitude", continuous_update=False
)

w_noise = widgets.FloatSlider(
    value=0.0, min=0.0, max=1.0, step=0.05,
    description="Bruit", continuous_update=False
)

out = widgets.interactive_output(
    plot_demo,
    {"freq": w_freq, "amplitude": w_amp, "noise": w_noise}
)

ui = widgets.VBox([w_freq, w_amp, w_noise, out])
display(ui)
```
