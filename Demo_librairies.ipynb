```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import FloatSlider, VBox, interactive_output
from IPython.display import display

def f(a=1.0):
    x = np.linspace(0, 10, 200)
    y = np.sin(a * x)
    fig, ax = plt.subplots(figsize=(6, 3))
    ax.plot(x, y)
    plt.show()

w = FloatSlider(value=1.0, min=0.5, max=5.0, step=0.1, description="a")
out = interactive_output(f, {"a": w})
ui = VBox([w, out])
display(ui)
```
